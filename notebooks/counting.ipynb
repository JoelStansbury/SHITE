```python
def percentUpper(s):
    n = 0
    N = len(s)
    for char in s:
        if char.isupper(): n += 1
    return n/N
```

```python
from functools import cache
```

```python
@cache
def choose(n, k):
    """
    A fast way to calculate binomial coefficients by Andrew Dalke (contrib).
    """
    if 0 <= k <= n:
        ntok = 1
        ktok = 1
        for t in range(1, min(k, n - k) + 1):
            ntok *= n
            ktok *= t
            n -= 1
        return ntok // ktok
    else:
        return 0
```

```python
def categorize(samples, thresh=0.001):
    '''
    assuming equally probable samples, identify samples which occur with
    count high enough that the probability of seeing that count
    given the "equal probability" assumtion is less than the threshold.
    
    thresh of 0.001 means there will be at least 99.9% confidence that 
    the categories returned are not random
    '''
    dd = defaultdict(int)
    for s in samples:
        dd[s] += 1
    n_cats = len(dd)
    p = 1/n_cats
    
    # calculate how many instances would constitute a 99.9% confidence
    N = len(samples)
    n = N
    p_ = p**n  # probability of N instances of a class in N samples
    while p_ < thresh:
        n-=1
        p_ += choose(N, n) * (p**n) * (1-p)**(N-n)  # probability of exactly n instances if a class
    
    good_cats = {
        k[0]:i 
        for i, k in enumerate(
            sorted(
                dd.items(), 
                key=lambda x: x[1],
                reverse=True
            )
        ) if k[1]>=n
    }
    cats = defaultdict(lambda:-1)
    cats.update(good_cats)
    return [cats[s] for s in samples]
    
```
